```python
# Cargar Modulos 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('iris')
data.head(5)
```

```python
# Resumen de datos
data.describe()
```

```python
# Distribución de cada clase
data.groupby('species').size()
```

```python
# Carga de bibilotecas especificas para modelos de IA
from sklearn.model_selection import train_test_split
```

```python
# Separacion de datos en grupo de Entrenamiento y Prueba 
# Random_state es para usar la misma semilla osea los datos se pueden replicar 
train, test = train_test_split(data, test_size=0.4, stratify=data['species'], random_state=19)
train.head(10)
```

```python
## Analissi Exploratorio 
```

```python
# Histroramas para observar la distribvución de los 4 atrbiutos
n_bins = 10
fig, axs = plt.subplots(2,2)
axs[0,0].hist(train['sepal_length'], bins=n_bins)
axs[0,0].set_title('Longitud de sepalo')
axs[0,1].hist(train['sepal_width'], bins=n_bins)
axs[0,1].set_title('Ancho de sepalo')
axs[1,0].hist(train['petal_length'], bins=n_bins)
axs[1,0].set_title('Longitud de Petalo')
axs[1,1].hist(train['petal_width'], bins=n_bins)
axs[1,1].set_title('Ancho de Petalo')
fig.tight_layout(pad=1.0);
```

```python
# Biagramas Box-plot
fn = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
cn = ['setosa', 'versicolor', 'virginica']
fig, axs = plt.subplots(2,2)
sns.boxplot(x='species', y=fn[0], data=train, order=cn, ax=axs[0,0])
sns.boxplot(x='species', y=fn[1], data=train, order=cn, ax=axs[0,1])
sns.boxplot(x='species', y=fn[2], data=train, order=cn, ax=axs[1,0])
sns.boxplot(x='species', y=fn[3], data=train, order=cn, ax=axs[1,1])
sns.set_palette("flare")
fig.tight_layout(pad=1.0);
```

```python
# Diagramas de Violin
sns.violinplot(x='species',y=fn[2], data=train, order=cn, palette='colorblind', legend=False);
```

```python
# Diagrama de dispersión de atributos emparejados
sns.pairplot(train, hue='species', height=2, palette='colorblind');
```

```python
pd.plotting.parallel_coordinates(train, 'species',color=['blue','red','green']);
```

```python
### separacion de la etiqueta de clase y atributos
```

```python
x_train = train[['sepal_length','sepal_width','petal_length','petal_width']]
y_train = train['species']

x_test = test[['sepal_length','sepal_width','petal_length','petal_width']]
y_test = test['species']

print(x_train.head(5))
print(y_train.head(5))
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
mod_dt = DecisionTreeClassifier(max_depth=3, random_state=1)
mod_dt.fit(x_train, y_train)
```

```python
### Visualización
```

```python
plt.figure(figsize=(10,8))
plot_tree(mod_dt, feature_names=mod_dt.feature_names_in_, class_names=cn, filled=True);
```

```python
prediccion = mod_dt.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print('la precisión del árbol de decisión es: {:.3f}'.format(metrics.accuracy_score(prediccion, y_test)))
```
